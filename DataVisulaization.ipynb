```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/iris.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='sepal_width',y='sepal_length',hue='species',style='species',palette='coolwarm')
plt.title('Sepal Lenght vs SepalWidth (Seaborn)')
plt.show()
```

```python
species_counts=df['species'].value_counts()
species_counts_df=species_counts.reset_index()
plt.figure(figsize=(10,6))
plt.bar(species_counts_df['species'],species_counts_df['count'],color='maroon')
plt.title('Bar plot of Iris Shape Species(Matplotlib)')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=species_counts_df,x='species',y='count',palette='magma')
plt.title('Bar Plot of Iris Species Seaborn')
```

```python
plt.figure(figsize=(10,6))
df.boxplot(column='sepal_length',by='species',grid=True)
plt.title('Box Plot of Sepal Length (Matplotlib)')
plt.suptitle('')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='species',y='sepal_length',hue=df['species'])
plt.title('Box plot of Sepal Length (Seaborn)')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

**Heatmap**

```python
correlation_matrix=df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='magma',fmt='.2f',square=True,xticklabels='12345')
plt.xticks(rotation=45,ha='right')
plt.title('Correlation Heatmap (Seaborn)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x='species',y='sepal_length',palette='coolwarm')
plt.title('vIOLIN PLOT OF SEPAL LENGTH (Seaborn)')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(axis='y')
plt.show()
```
