# **Lab Objective:**
1. To simulate a population distribution (e.g., uniform or exponential distribution).

2. To take multiple random samples of a given size from this population and compute their means.

3. To observe how the distribution of these sample means approaches a normal distribution as the number of samples and sample size increase.


# **Required Libraries:**
1. numpy: For data manipulation and random sampling.

2. matplotlib: For visualization.

3. seaborn: For enhanced visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(60)
#set random seed for reproducilbility
#step 1 : define a population distributoin ( non -normal distribution )
population_size=100000
# step 2: make sample sizes to demonstrate convergence
sample_sizes=[5,10,30,50,70,80,100]
#step 3: take number of samples for each sample size .
num_samples=1000
population=np.random.exponential(scale=2,size=population_size)
#here , population_size=100000, this will generate 100000 random samples from
# exponential distribution with a mean of 2
print(population)
# print(population.mean())
#Exponential distribution
plt.figure(figsize=(10,6))
sns.histplot(population,bins=50,kde=True,color="green",alpha=0.7)
plt.title("Exponential Population Distribution",fontsize=15)
plt.xlabel("Value",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

Now, taking out the samples from the exponential distribution and check whether the distribution is normal distribution curve or not.

```python
plt.figure(figsize=(20,16))
for i,size in enumerate(sample_sizes,start=1):
  sample_means=[np.mean(np.random.choice(population,size=size,replace=False)) for _ in range(num_samples)]
  print(i)
  print(size)
  print(sample_means)
  #plot the distribution of sample means
  plt.subplot(3 ,2 ,i)
  sns.histplot(sample_means,bins=30,kde=True,color="orange",alpha=0.7)
  plt.title(f"Sample Size={size}",fontsize=15)
  plt.xlabel("Sample Mean",fontsize=15)
  plt.ylabel("Frequency",fontsize=15)

#show the final plots
plt.suptitle("Demonstration of the central limit theorem ",fontsize=18, y =1.08)
plt.tight_layout()
plt.show()
```

As sample size increases, the sample means' distribution converges to a normal distribution, regardless of the population's original distribution.
